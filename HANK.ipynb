# Solving a one-asset HANK model in sequence space
In this notebook, we show how to

1. Calculate the GE of a HANK model and calibrate its steady state
2. Obtain transitional dynamics from MIT shocks using the sequence space Jacobian.
3. Simulate an economy hit by aggregate shocks using the BKM (2018) algorithm. 

The model is a HA version of the textbook New Keynesian model. Technology is linear in labor, prices are sticky á la Rotemberg, and monetary policy follows a Taylor rule. For a detailed description of the model, please see Auclert et al. (2020): "Using the Sequence-Space Jacobian to Solve and Estimate Heterogeneous Agent Models". The basic building blokcs of the model are as follows:

$$
\textbf{H}_t(\textbf{Y}, Z) \equiv 
\begin{pmatrix}
Y_t - Z_t L_t
\\
Y_t \left[1 - \frac{\mu}{\mu-1}\frac{1}{2\kappa} \log(1 + \pi_t)^2\right] - w_t L_t - d_t
\\
r_t B - \tau_t
\\
r^*_t + \phi \pi_t - i_t
\\
1 + r_t - \frac{1+i_{t-1}}{1+\pi_t}
\\
\kappa \left(\frac{w_t}{Z_t} - \frac{1}{\mu} \right)  + \frac{1}{1+r_{t+1}} \frac{Y_{t+1}}{Y_t} \log(1+\pi_{t+1}) - \log(1+\pi_t)
\\
\mathcal{A}_t(\{r_s, w_s, \tau_s, d_s\})  - B
\\
\mathcal{N}_t(\{r_s, w_s, \tau_s, d_s\}) - L_t
\end{pmatrix}
= \begin{pmatrix} 0 \\ 0 \\ 0 \\ 0 \\ 0 \\ 0 \\ 0 \\ 0\end{pmatrix},
\qquad t = 0, 1, \dots
$$

where the endogenous variables are $\textbf{Y} = (Y, L, r, w, d, \pi, \tau, i)$ and the exogenous variables are $\textbf{Z}=(r^*, Z)$. 

The asset demand and labor supply functions $\{\mathcal{A}, \mathcal{L}\}$ follow from the household block with Bellman equation

$$
\begin{align} \tag{HH}
V_t(e, a_{-}) = \max_{c, n, a} &\left\{\frac{c^{1-\sigma}}{1-\sigma} - \varphi \frac{n^{1+\nu}}{1+\nu} + \beta \sum_{e'}V_{t+1}(e', a) \mathcal{P}(e,e') \right\}
\\
c + a &= (1 + r_t)a_{-} + w_t e n - \tau_t \bar{\tau}(e) + d_t \bar{d}(e)
\\
a &\geq 0
\end{align}
$$

where $\bar\tau(e)$ and $\bar d(e)$ are skill-specific incidence rules for taxes and dividends. 

To see the directed acyclic graph (DAG) that accompanies sequence-space models, please refer to the paper.

## Step 1: Solving for the steady-state using EGM

The main function "egm" will complete all the backward iterations we need when solving for the jacobian in the household block.

```julia
include("rowenhorst_tauchen.jl") # include discretization functions (make sure it is in your working directory)

using Parameters, Plots, LinearAlgebra, Interpolations, BenchmarkTools, Roots, QuantEcon

Household = @with_kw (apoints = 500, #asset grid
    amax = 150,  # asset max
    beta = 0.986, # discount factor
    alpha = 0.11, # capital share
    deprec = 0.025, # depreciation rate
    gamma = 2.0, # inverse elasticity of substitution
    bc = 0, # borrowing constraint (must be weakly negative)
    rho = 0.966,  # autocorr of income process
    uncond_sd = 0.5, # unconditional sd
    num_states = 7, # number of states for income process
    sd = sqrt(1-rho^2)*uncond_sd, # stand. dev. of deviation process
    mean = 0, # mean of income process
    transition_matrix = rowenhorst(mean, uncond_sd, rho, num_states)[1], # transition matrix
    stationary_dis = stationary_distributions(MarkovChain(transition_matrix))[1], # stationary dis of income process
    ygrid = rowenhorst(mean, uncond_sd, rho, num_states)[2], # grid for income process
    Amat = setgrids(bc, apoints, amax, num_states, ygrid)[1], # asset grid
    Ymat = setgrids(bc, apoints, amax, num_states, ygrid)[2], # income grid
    phi = 0.8, # disutility factor
    frisch = 2.0, # inverse frisch elasticity for labor supply
    mu = 1.2, # markup
    B_Y = 5.6, # liquid assets
    r = 0.005) # steady state interest rate)


function setgrids(bc, apoints, amax, num_states, ygrid)
    Amat = [i for i in range(bc, length = apoints, stop= amax), j in 1:num_states]
    Ymat = [j for i = 1:apoints, j in ygrid]
    return Amat, Ymat
end

function setunevengrids(bc, apoints, amax, num_states, ygrid; curve = 2)
    unevengrid = range(bc^(1/curve), stop = amax^(1/curve), length = apoints).^curve
    Amat = [i for i in unevengrid, j in 1:num_states]
    Ymat = [j for i = 1:apoints, j in ygrid]
    return Amat, Ymat
end

hh = Household()
```

```julia
function transfers(stationary_dis; div_rule, tax_rule, Div, Tax)
    div = Div / sum(stationary_dis .* div_rule) .* div_rule
    tax = Tax / sum(stationary_dis .* tax_rule) .* tax_rule
    T = div .- tax
    return T
end
```

```julia
up(c,gamma) = c.^(-gamma)
invup(x,gamma) = x.^(-1/gamma)
vp(l, phi, frisch) = phi.*l.^frisch
invvp(x, phi, frisch) = (x./phi).^(1/frisch)

# euler equation to get current consumption given future interest rate and future consumption grid

function getc(gamma,beta,transition_matrix; rnext,cnext)
    upnext = beta.*(1+rnext).*up(cnext,gamma)*transition_matrix' # future marginal utility
    c = invup(upnext,gamma) # current consumption
    return c
end

# intratemporal foc to get labor given current consumption

function getl(c, y, gamma, phi, frisch, w)
    labor = invvp(w.*y.*(up(c, gamma)), phi, frisch)
    return labor
end

# obtain current assets given consumption today defined on asset grid tomorrow

function geta(Amat, Ymat, gamma, phi, frisch, T; r, w, c)
    return 1/(1+r).*(c.+Amat.-w.*Ymat.*getl(c, Ymat, gamma, phi, frisch, w) .- T)
end

# obtain aggregates

function getagg(; policyfun, dist)
    agg = reshape(policyfun, (length(dist),1))
    agg = sum(agg.*dist)
    return agg
end
```

```julia
# Note: function takes both current and future interest rates as arguments.

function egm(hh; w, cnext, cbinding, r, rnext, T)
 """
    use endogenous grid method to obtain c_{t} and a_{t} given c_{t+1} 'cnext'
    
    #### Fields
    
    - 'hh': household tuple
    - 'w': wage rate
    - 'cnext': time t+1 consumption grid
    - 'cbinding': consumption grid when borrowing constraint binds
    - 'r': interest rate at time t
    - 'rnext': interest rate at time t+1
    
    #### Returns
    
    - 'c': time t consumption grid
    - 'anext': time t policy function
    - 'l': time t labor function
"""
    
@unpack gamma, beta, transition_matrix, stationary_dis, Amat, Ymat, num_states, phi, frisch, bc = hh

# current policy functions on future grid
c = getc(gamma, beta, transition_matrix; rnext=rnext, cnext=cnext)
a = geta(Amat, Ymat, gamma, phi, frisch, T; r=r, w=w, c=c)

cnonbinding = similar(Amat)
    
# get consumption policy function for current grid
for i = 1:num_states
    cnonbinding[:,i] = LinearInterpolation(a[:,i], c[:,i], extrapolation_bc = Line()).(Amat[:,i])
end


# update elements of consumption policy when borrowing constraint binds
# a[1,j] is the level of current assets that induces the borrowing constraint to bind exactly.
# Therefore, whenever current assets are below a[1,j], the borrowing constraint will be STRICTLY binding.
# Note that this uses the monotonicity of the policy rule.

for j = 1:num_states
   c[:,j] = (Amat[:,j] .> a[1,j]) .*cnonbinding[:,j] .+ (Amat[:,j] .<= a[1,j]).*cbinding[:,j]
end
    
l = getl(c, Ymat, gamma, phi, frisch, w)
    
# update saving policy function with new consumption function
anext = @. (1+r)*Amat + w*Ymat*l - c + T

return c, anext, l
end
```

```julia
function egm_iterate(hh; r, w, T, rnext, cnext)
    @unpack Amat, Ymat, bc, gamma, phi, frisch, num_states, apoints = hh
    
    function getcbinding(a, c, y, t)
        @. (1+r)*a + w*y*getl(c, y, gamma, phi, frisch, w) + t - bc - c
    end
    
    cbinding = similar(Amat)
    
    for j = 1:num_states
        for i = 1:apoints
            cbinding[i,j] = find_zero(c -> getcbinding(Amat[i,j], c, Ymat[i,j], T[1,j]), 1.0)
        end
    end
    
    c, anext, l = egm(hh; w=w, rnext=rnext, r=r, T = T, cnext=cnext, cbinding=cbinding)
    
    return c, anext, l
end

function hh_ss(hh; r, w, T, tol=1e-8, maxiter=1000)
    @unpack Amat, Ymat, bc, gamma, phi, frisch, num_states, apoints = hh
    
    cnext = @. (max(r,0))*Amat+Ymat*w + T # initial guess for policy function iteration 
    
    # get consumption when borrowing constraint binds
    
    function getcbinding(a, c, y, t)
        @. (1+r)*a + w*y*getl(c, y, gamma, phi, frisch, w) + t - bc - c
    end
    
    cbinding = similar(Amat)
    
    for j = 1:num_states
        for i = 1:apoints
            cbinding[i,j] = find_zero(c -> getcbinding(Amat[i,j], c, Ymat[i,j], T[1,j]), 1.0)
        end
    end
    
    dist = 1
    counter = 0
    
    for i=1:maxiter
        c = egm(hh; w=w, rnext=r, r=r, T = T, cnext=cnext, cbinding=cbinding)[1]
        if norm(c-cnext,Inf)<tol 
            println("policy function iteration successful!")
            return egm(hh; w=w, rnext=r, r=r, T = T, cnext=c, cbinding=cbinding)
        else
            cnext = c
        end
    end
    
    error("no policy function convergence!")
    
end
```

```julia
function get_kernel(hh; policyfun)
    
"""
    get transition kernel for asset-income distribution using Young's method
    
    #### Fields
    
    - 'hh': household tuple
    - 'policyfun': savings function, array na x ny
    
    #### Returns
    
    - 'kernel': na*ny x na*ny array row-stochastic transition kernel
    
"""

@unpack apoints, Amat, num_states, Ymat, transition_matrix = hh
    
# construct transition matrix without labour stochasticity
Q = zeros(apoints, apoints, num_states)
    
# find which element to interpolate to
findnearest(A,t) = findmin(abs.(A.-t))[2]

# construct matrices to get neighboring indices
next = similar(Amat)
previous = similar(Amat)

for j = 1:num_states
    for k = 1:apoints
        ind = findnearest(Amat[:,j],policyfun[k,j])
            if policyfun[k,j] - Amat[ind,j] > 0
                previous[k,j] = Amat[ind,j]
                if ind == apoints
                    next[k,j] = Amat[ind,j]
                else         
                next[k,j] = Amat[ind+1,j]
                end
            elseif policyfun[k,j] == Amat[ind,j]
                previous[k,j] = policyfun[k,j] 
                next[k,j] = policyfun[k,]
            else
                next[k,j] = Amat[ind,j]
                if ind == 1
                    previous[k,j] = Amat[ind,j]
                else
                    previous[k,j] = Amat[ind-1,j]
                end
            end  
    end
end
    
# construct transition matrix for assets.
# Q[i,j,k] is the probability of going to agrid[i] when current assets are agrid[j] and income is ygrid[k]
for k = 1:apoints
    for j = 1:num_states
        if next[k,j] == previous[k,j]
            Q[Amat[:,1].==previous[k,j],k,j] .= 1
        else
            Q[Amat[:,1].==previous[k,j],k,j] .= (next[k,j]-policyfun[k,j])./(next[k,j]-previous[k,j])
            Q[Amat[:,1].==next[k,j],k,j] .= 1 - (next[k,j]-policyfun[k,j])./(next[k,j]-previous[k,j])  
        end
    end
end
    
# construct transition kernel by taking the kronecker product of assets and transition_matrix for income
kernel = zeros(apoints*num_states,apoints*num_states)

for j=1:num_states
    for i=1:num_states
        kernel[(i-1)*apoints+1:i*apoints,(j-1)*apoints+1:j*apoints]=Q[:,:,j].*transition_matrix[j,i];
    end
end

kernel = kernel'
    
return kernel
end
```

```julia
function histogram_method(hh; kernel, tol=1e-8, maxiter=10000)
    #needs Markov transition kernel
    @unpack apoints, num_states = hh
    
    #initial guess for invariant distribution
    init = ones(apoints*num_states)
    init = init./sum(init)
    
    dist = 1
    counter = 0
    
    for i=1:maxiter
        initnext = (init'*kernel)'
        if norm(init-initnext,Inf) < tol 
            println("distribution iteration successful!")
            return initnext
        else
            init = initnext
        end
    end
    
    error("distribution did not convergence!")
end
```

```julia
function get_hh_agg(hh, r, w, T)
    @unpack beta, Ymat = hh
    
    c, anext, l = hh_ss(hh; r = r, w = w, T = T)
    
    kernel = get_kernel(hh; policyfun = anext) # get kernel
    invariant_distribution = histogram_method(hh; kernel = kernel) # get invariant distribution
    
    K = getagg(; policyfun = anext, dist = invariant_distribution) # get aggregate capital
    L = getagg(; policyfun = l.*Ymat, dist = invariant_distribution) # get aggregate labor
    return K, L
end

function get_hh_agg(hh; c, anext, l, dist)
    @unpack Ymat = hh
    
    K = getagg(; policyfun = anext, dist = dist) # get aggregate capital
    L = getagg(; policyfun = l.*Ymat, dist = dist) # get aggregate labor
    C = getagg(; policyfun = c, dist = dist)
    return C, K, L
end
```

```julia
function obtain_J(f, x, y; h = 1e-5)
    """Finds Jacobian f'(x) around y=f(x)"""
    
    nx = size(x)[1]
    ny = size(y)[1]
    J = zeros(ny, nx)

    for i = 1:nx
        dx = h * (1:nx .== i)
        J[i, :] = (f(x .+ dx) .- y) ./ h
    end
    return J
end

function newton_solver(f, x, y; tol = 1e-6, maxcount = 100)
    for count=1:maxcount
        J = obtain_J(f, x, y)
        
        if maximum(abs.(y)) < tol
            return x, y
        end
    
        dx = J\(-y)
        print(dx)
        y = f(x + dx)
        x = x + dx
        print(y)
    end
    error("no convergence in newton solver")
end
```

## Step 2: Calibrating the Steady-State

We use a Newton method to calibrate the equilibrium bond holdings and equilibrium output in steady-state. This will generally be faster than ad hoc methods that use a variant of the bisection method.



```julia
function calibrate_ss!(hh; mu = 1.2, r = 0.005, B_Y = 5.6)
    div_rule = hh.ygrid
    tax_rule = hh.ygrid
    
    # solve analytically what we can
    B = B_Y
    w = 1 / mu
    Div = (1 - w)
    Tax = r * B
    T = transfers(hh.stationary_dis; div_rule, tax_rule, Div, Tax)'
    
    x = [0.982041, 0.786046] # guess for beta and phi
    temp_hh = merge(hh, (beta = x[1], phi = x[2]))
    y = [get_hh_agg(temp_hh, r, w, T)[1] - B; get_hh_agg(temp_hh, r, w, T)[2] - 1]
    
    function residual(x)
       temp_hh = merge(hh, (beta = x[1], phi = x[2]))
       return [get_hh_agg(temp_hh, r, w, T)[1] - B; get_hh_agg(temp_hh, r, w, T)[2] - 1]
    end
    
    result = newton_solver(residual, x, y)[1]
    
    hh = merge(hh, (beta = result[1], phi = result[2]))
    
    print(get_hh_agg(hh, r, w, T)[1])
    
    return hh
end

hh = calibrate_ss!(hh)
```

## Step 3: Getting the household block jacobian.

The following set of functions calculate the jacobians of the "outputs" of the household block with respect to certain inputs. This is the most time consuming part of the sequence-space jacobian approach. The code utilizes the **fake news** algorithm, which significantly speeds up the calculation of the jacobians. Please read the paper to obtain familiarity with some of the terminology used in the code.


```julia
#getting jacobian with respect to r

function get_dYr(hh; horizon, diff, css, dist)

    @unpack B_Y, mu, r, Amat = hh
    
    div_rule = hh.ygrid
    tax_rule = hh.ygrid
    
    # solve analytically what we can
    B = B_Y
    w = 1 / mu
    Div = (1 - w)
    Tax = r * B
    T = transfers(hh.stationary_dis; div_rule, tax_rule, Div, Tax)'
    
    r_up = r + diff
    r_down = r - diff


    
#vector that holds aggregate differences

dK = zeros(horizon)
dL = zeros(horizon)
dC = zeros(horizon)
    
# vector of one-dimensional vectors (must be column vectors!) to hold change in distributions
    
dD = zeros(length(Amat),horizon)
    
#first upward iteration
cup, aup, lup = egm_iterate(hh; r = r_up, w = w, T = T, rnext = r, cnext = css)

Λ_up = get_kernel(hh; policyfun = aup) #get transition function (up)

#first downward iteration
cdown, adown, ldown = egm_iterate(hh; r = r_down, w = w, T = T, rnext = r , cnext = css)

Λ_down = get_kernel(hh, policyfun = adown) #get transition function (down)

#get capital and consumption
    
dC[1], dK[1], dL[1] = get_hh_agg(hh; c = (cup.-cdown)./2, anext = (aup.-adown)./2, l = (lup.-ldown)./2, dist = dist)

#get change in distribution
dD[:,1] = (dist'*((Λ_up .- Λ_down)./2))'
    
#second upward iteration 
cup, aup, lup = egm_iterate(hh; r = r, w = w, T = T, rnext = r_up, cnext = cup)
    
Λ_up = get_kernel(hh; policyfun = aup) #get transition function (up)

#second downward iteration
cdown, adown, ldown = egm_iterate(hh; r = r, w = w, T = T, rnext = r_down, cnext = cdown)
    
Λ_down = get_kernel(hh; policyfun = adown) #get transition function (down)
    
#get capital and consumption
dC[2], dK[2], dL[2] = get_hh_agg(hh; c = (cup.-cdown)./2, anext = (aup.-adown)./2, l = (lup.-ldown)./2, dist = dist)


#change in distribution
dD[:,2] = (dist'*((Λ_up .- Λ_down)./2))'

for i=1:horizon-2
    #upward iteration
    cup, aup, lup = egm_iterate(hh; r = r, w = w, T = T, rnext = r, cnext = cup)
    Λ_up = get_kernel(hh; policyfun = aup)
    
    #downward iteration
    cdown, adown, ldown = egm_iterate(hh; r = r, w = w, T = T, rnext = r, cnext = cdown)
    Λ_down = get_kernel(hh; policyfun = adown) #get transition function (down)
    
    #get capital, consumption, and distribution
    dC[i+2], dK[i+2], dL[i+2] = get_hh_agg(hh; c = (cup.-cdown)./2, anext = (aup.-adown)./2, l = (lup.-ldown)./2, dist = dist)

    dD[:,i+2] = (dist'*((Λ_up .- Λ_down)./2))'
end     
    
    return dK./diff, dC./diff, dL./diff, dD./diff

end
```

```julia
# getting jacobian with respect to w

function get_dYw(hh; horizon, diff, css, dist)

    @unpack B_Y, mu, r, Amat = hh
    
    div_rule = hh.ygrid
    tax_rule = hh.ygrid
    
    # solve analytically what we can
    B = B_Y
    w = 1 / mu
    Div = (1 - w)
    Tax = r * B
    T = transfers(hh.stationary_dis; div_rule, tax_rule, Div, Tax)'
    
    w_up = w + diff
    w_down = w - diff


    
#vector that holds aggregate differences

dK = zeros(horizon)
dL = zeros(horizon)
dC = zeros(horizon)
    
# vector of one-dimensional vectors (must be column vectors!) to hold change in distributions
    
dD = zeros(length(Amat),horizon)
    
#first upward iteration
cup, aup, lup = egm_iterate(hh; r = r, w = w_up, T = T, rnext = r, cnext = css)

Λ_up = get_kernel(hh; policyfun = aup) #get transition function (up)

#first downward iteration
cdown, adown, ldown = egm_iterate(hh; r = r, w = w_down, T = T, rnext = r , cnext = css)

Λ_down = get_kernel(hh, policyfun = adown) #get transition function (down)

#get capital and consumption
    
dC[1], dK[1], dL[1] = get_hh_agg(hh; c = (cup.-cdown)./2, anext = (aup.-adown)./2, l = (lup.-ldown)./2, dist = dist)

#get change in distribution
dD[:,1] = (dist'*((Λ_up .- Λ_down)./2))'
    

for i=1:horizon-1
    #upward iteration
    cup, aup, lup = egm_iterate(hh; r = r, w = w, T = T, rnext = r, cnext = cup)
    Λ_up = get_kernel(hh; policyfun = aup)
    
    #downward iteration
    cdown, adown, ldown = egm_iterate(hh; r = r, w = w, T = T, rnext = r, cnext = cdown)
    Λ_down = get_kernel(hh; policyfun = adown) #get transition function (down)
    
    #get capital, consumption, and distribution
    dC[i+1], dK[i+1], dL[i+1] = get_hh_agg(hh; c = (cup.-cdown)./2, anext = (aup.-adown)./2, l = (lup.-ldown)./2, dist = dist)

    dD[:,i+1] = (dist'*((Λ_up .- Λ_down)./2))'
end     
    
    return dK./diff, dC./diff, dL./diff, dD./diff

end
```

```julia
# getting jacobian with respect to T

function get_dYT(hh; horizon, diff, css, dist)

    @unpack B_Y, mu, r, Amat = hh
    
    div_rule = hh.ygrid
    tax_rule = hh.ygrid
    
    # solve analytically what we can
    B = B_Y
    w = 1 / mu
    Div = (1 - w)
    Tax = r * B
    T = transfers(hh.stationary_dis; div_rule, tax_rule, Div, Tax)'
    
    T_up = T .+ diff
    T_down = T .- diff


    
#vector that holds aggregate differences

dK = zeros(horizon)
dL = zeros(horizon)
dC = zeros(horizon)
    
# vector of one-dimensional vectors (must be column vectors!) to hold change in distributions
    
dD = zeros(length(Amat),horizon)
    
#first upward iteration
cup, aup, lup = egm_iterate(hh; r = r, w = w, T = T_up, rnext = r, cnext = css)

Λ_up = get_kernel(hh; policyfun = aup) #get transition function (up)

#first downward iteration
cdown, adown, ldown = egm_iterate(hh; r = r, w = w, T = T_down, rnext = r , cnext = css)

Λ_down = get_kernel(hh, policyfun = adown) #get transition function (down)

#get capital and consumption
    
dC[1], dK[1], dL[1] = get_hh_agg(hh; c = (cup.-cdown)./2, anext = (aup.-adown)./2, l = (lup.-ldown)./2, dist = dist)

#get change in distribution
dD[:,1] = (dist'*((Λ_up .- Λ_down)./2))'
    

for i=1:horizon-1
    #upward iteration
    cup, aup, lup = egm_iterate(hh; r = r, w = w, T = T, rnext = r, cnext = cup)
    Λ_up = get_kernel(hh; policyfun = aup)
    
    #downward iteration
    cdown, adown, ldown = egm_iterate(hh; r = r, w = w, T = T, rnext = r, cnext = cdown)
    Λ_down = get_kernel(hh; policyfun = adown) #get transition function (down)
    
    #get capital, consumption, and distribution
    dC[i+1], dK[i+1], dL[i+1] = get_hh_agg(hh; c = (cup.-cdown)./2, anext = (aup.-adown)./2, l = (lup.-ldown)./2, dist = dist)

    dD[:,i+1] = (dist'*((Λ_up .- Λ_down)./2))'
end     
    
    return dK./diff, dC./diff, dL./diff, dD./diff

end
```

```julia
# getting jacobian with respect to w

function get_dYbeta(hh; horizon, diff, css, dist)

    @unpack B_Y, mu, r, Amat, beta = hh
    
    div_rule = hh.ygrid
    tax_rule = hh.ygrid
    
    # solve analytically what we can
    B = B_Y
    w = 1 / mu
    Div = (1 - w)
    Tax = r * B
    T = transfers(hh.stationary_dis; div_rule, tax_rule, Div, Tax)'
    
    beta_up = beta + diff
    beta_down = beta - diff

hh_temp = merge(hh, (beta = beta_up,))
    
#vector that holds aggregate differences

dK = zeros(horizon)
dL = zeros(horizon)
dC = zeros(horizon)
    
# vector of one-dimensional vectors (must be column vectors!) to hold change in distributions
    
dD = zeros(length(Amat),horizon)
    
#first upward iteration
cup, aup, lup = egm_iterate(hh_temp; r = r, w = w, T = T, rnext = r, cnext = css)

Λ_up = get_kernel(hh_temp; policyfun = aup) #get transition function (up)

#first downward iteration
hh_temp = merge(hh, (beta = beta_down,))
cdown, adown, ldown = egm_iterate(hh_temp; r = r, w = w, T = T, rnext = r , cnext = css)

Λ_down = get_kernel(hh_temp, policyfun = adown) #get transition function (down)

#get capital and consumption
    
dC[1], dK[1], dL[1] = get_hh_agg(hh; c = (cup.-cdown)./2, anext = (aup.-adown)./2, l = (lup.-ldown)./2, dist = dist)

#get change in distribution
dD[:,1] = (dist'*((Λ_up .- Λ_down)./2))'
    

for i=1:horizon-1
    #upward iteration
    cup, aup, lup = egm_iterate(hh; r = r, w = w, T = T, rnext = r, cnext = cup)
    Λ_up = get_kernel(hh; policyfun = aup)
    
    #downward iteration
    cdown, adown, ldown = egm_iterate(hh; r = r, w = w, T = T, rnext = r, cnext = cdown)
    Λ_down = get_kernel(hh; policyfun = adown) #get transition function (down)
    
    #get capital, consumption, and distribution
    dC[i+1], dK[i+1], dL[i+1] = get_hh_agg(hh; c = (cup.-cdown)./2, anext = (aup.-adown)./2, l = (lup.-ldown)./2, dist = dist)

    dD[:,i+1] = (dist'*((Λ_up .- Λ_down)./2))'
end     
    
    return dK./diff, dC./diff, dL./diff, dD./diff

end
```

### Expectation Vectors

```julia
function expect_vec(hh; policyfun, horizon, output)
    
    @unpack Amat = hh
    
    expv = zeros(length(Amat),horizon)

    trans= get_kernel(hh; policyfun = policyfun)
    vec = reshape(output, length(trans[:,1]))
    
    expv[:,1] = vec
    
    for i=2:horizon-1
       expv[:,i] = (trans)*expv[:,i-1]
    end
    
    return expv
end

function fake_news(;horizon, expv, dD, dY)
    #constructs fake news matrix
    #inputs
#         expv: vector of expectation vectors
#         dD: vector of change in distribution (vector of vectors again)
#         dY: changes in outputs, vector of floats
   
    F = zeros(horizon,horizon)
    
    for j=1:horizon
        for i=1:horizon
            if i == 1
                F[i,j] = dY[j]
            else
                F[i,j] = expv[:,i-1]'*dD[:,j]
            end
        end
    end
    
    return F
end

function get_jac(F)
    #constructs Jacobian
   J = copy(F)
    for j = 2:length(F[:,1])
        for i = 2:length(F[1,:])
            J[i,j] = J[i-1,j-1] + F[i,j]
        end
    end
    
    return J
end

function GET_JAC(hh; policyfun, output, horizon, dD, dY)
    expv = expect_vec(hh; policyfun = policyfun, horizon = horizon, output = output)
    F = fake_news(horizon = horizon, expv = expv, dD = dD, dY = dY)
    J = get_jac(F)
    return J
end
```

## Step 4: Calculating the Jacobians with respect to all other blocks

In these relatively large models, it is very error-prone to calculate the Jacobians for each block by hand, "accumulate" them until the market clearing block, and then use the implicit function theorem. You'll be dealing with a lot of jacobians ($TxT$ matrices) -- one for each output and input for **every** block -- and you need to make sure to multiply them in the right order. 

I suggest doing the computationally heavy parts in Julia (this will generally be calculating the jacobian of the household block) and then using Auclert et al.'s code to do the computationally simpler -- but tedious -- work of calculating the general equilibrium jacobians. ## Step 6: Simulating an economy with the BKM algorithm

Finally, we can use the BKM algorithm to simulate an economy hit by aggregate shocks. The simulation will be accurate as long as the aggregate shocks are small, or the model is "sufficiently" linear (not financial accelerator models, for example).

We can use these simulations to calculate standard deviations of various variables of interest, as well as cross-correlations. This is observable data that can be used to calibrate certain parameters (such as the standard deviations of the driving productivity process, or the persistence of the discount factor shock).

```julia
# BKM

using Distributions

YZ = reshape(df[!,"YZ"], (horizon, horizon))
rho = 0.7
dZ = rho.^(0:horizon-1)
Z_irf = YZ*dZ

Ybeta = reshape(df[!,"Ybeta"], (horizon, horizon))
rho = 0.7
dbeta = rho.^(0:horizon-1)
beta_irf = Ybeta*dbeta

sigma_z = 0.5
sigma_beta = 0.1

sz = Normal(0,sigma_z)
sb = Normal(0, sigma_beta)

t_shocks = horizon

z_shocks = zeros(t_shocks)
b_shocks = zeros(t_shocks)

z_shocks[1] = rand(sz)
b_shocks[1] = rand(sb)

for i = 2:t_shocks
    z_shocks[i] = rho*z_shocks[i-1] + rand(sz)
    b_shocks[i] = rho*b_shocks[i-1] + rand(sb)
end

output = zeros(t_shocks)

for i=1:t_shocks
    for k=1:i
        output[i]+=Z_irf[k]*z_shocks[i-k+1]
    end
    for k=1:i
        output[i]+=beta_irf[k]*b_shocks[i-k+1]
    end
end

plot(output)
title!("output with productivity and discount factor shocks (BKM)" , legend = false)
xlabel!("quarters")
ylabel!("pp deviation from steady state")

```
